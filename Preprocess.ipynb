```python
from langdetect import detect
import openpyxl 
from openpyxl import Workbook
import datetime
```

```python
book = openpyxl.load_workbook('2019.xlsx')
sheet = book.active
counter = 1
index = 1
detectLanguage = ''
previousText = ''
```

```python
lastrow = 'B' + str(sheet.max_row)
cells = sheet['A1': lastrow]
```

```python
wb = Workbook()
ws = wb.active
```

```python
for c1, c2 in cells:
    if(index % 1000 == 0):
        datestring = str(datetime.datetime.now().time())
        print(datestring ,' - ',index)
    index += 1
    stringExcel = str(c2.value)
    try:
        detectLanguage = detect(stringExcel)
    except:
        print("Error in ",index)
    if(detectLanguage == 'en' and previousText != stringExcel and len(stringExcel.split()) > 4):
        ws.cell(counter, 1, c1.value)
        ws.cell(counter, 2, stringExcel)
        counter += 1
    previousText = stringExcel
wb.save('pre-2019.xlsx')
```
